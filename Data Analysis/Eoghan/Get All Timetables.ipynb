## Prepend Missing Header to ensure uniformity

```python
def line_prepender(filename, line):
    with open(filename, 'r+') as f:
        content = f.read()
        f.seek(0, 0)
        f.write(line.rstrip('\r\n') + '\n' + content)
        f.close()
```

# Looper

```python
master = pd.DataFrame()

for filename in os.listdir('bus_data/route_schedules_backup'):
    
    line, day, direction = filename.split("_")
    direction = direction.replace(".csv","")

    # read data from csv
    df = pd.read_csv("bus_data/route_schedules_backup/" + filename, low_memory=False,)

    df["line"] = line
    df["day"] = day
    df["direction"] = direction
    
    master = pd.concat([master, df])
    print(filename, "done")
    
master.to_csv("bus_data/static_data/timetables_all.csv")
```
