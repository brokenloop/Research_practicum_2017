```python
def extract_schedule(filepath, master):
        
    # read data from csv
    df = pd.read_csv(filepath, low_memory=False, header=None)
    df.columns = ["departure", "variant"]

    df["line"] = line
    df["day"] = day
    df["direction"] = direction

    return pd.concat([master, df], ignore_index=True)
    
    
    
```

```python
def main(directory):
    master = pd.DataFrame()
    
    cwd = os.getcwd()
    # print(cwd)
    for filename in os.listdir(cwd+"\\bus_data\\route_schedules"):
        filepath = cwd+"\\bus_data\\route_schedules\\"+filename
        line, day, direction = filename.split("_")
        direction = direction.replace(".csv","") 
        print("Opening file", filename)
        master = extract_schedule(filepath, master)
    print("Finished!")
 
    return master

```
