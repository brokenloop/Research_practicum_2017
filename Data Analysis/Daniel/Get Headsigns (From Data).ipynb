```python
import pandas as pd
import os


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("bus_data/static_data3/route_stops_all.csv")
stops = pd.read_csv("bus_data/static_data3/stops.csv")
```

```python
df.head()
```

```python
merge_df = pd.merge(df, stops, how='left' , on=['StopID'])
```

```python
df.shape
```

```python
merge_df.shape
```

```python
merge_df.head(1)
```

```python
def add_headsign(row, headsigns):
    return headsigns[row.LineID, row.JourneyPatternID]
```

```python
headsigns = {}

merge_df = merge_df.sort_values(['LineID', 'JourneyPatternID', 'Order'])

lastlines = merge_df.groupby(['LineID', 'JourneyPatternID']).tail(1)


for index, row in lastlines.iterrows():
        headsigns[row.LineID, row.JourneyPatternID] = row['Name']
        
merge_df['Headsign'] = ""

merge_df['Headsign'] = merge_df.apply(lambda row: add_headsign(row, headsigns),axis=1)



# df['Runtime'] = df.apply(lambda row: add_headsign(row, headsigns),axis=1)

        
```

```python
final_df = merge_df[['LineID', 'JourneyPatternID', 'Headsign']].drop_duplicates()
```

```python
final_df.shape
```

```python
final_df.to_csv("bus_data/static_data3/routes_all(headsigns).csv", index=False)
```
